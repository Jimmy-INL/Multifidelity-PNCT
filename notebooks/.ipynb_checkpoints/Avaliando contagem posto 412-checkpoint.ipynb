```python
%matplotlib inline  

import os

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# Estatísticas descritivas
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# GP framework
import GPy

# Modelos clássicos
from statsmodels.tsa import stattools
from statsmodels import tsa

# sklearn GP
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import *
```

```python
postos = ['412']
anos = ['2016', '2017']

df = pd.Series()
for posto in postos:
    for ano in anos:

        df_posto = pd.read_pickle(f'../data/{posto}_{ano}_Imputado.pkl.gz', compression='gzip')
        df_posto = df_posto.sort_values(by='Data', ascending=True)
        df_posto['dia_data'] = df_posto.Data.dt.floor('d')
        
        df_contagem = df_posto.groupby('dia_data').size().reset_index()
        df = pd.concat([df, df_contagem])
```

```python
df = df.reset_index().drop(columns='index')
```

```python
df.head()
```

```python
df = df[df.columns[::-1]]
df.columns = ['Data', 'Contagem']
```

```python
df.head()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=df, x='Data', y='Contagem');
```

01-01-2016 Sexta feira

```python
mapa_dia_semana = {
    0: 'Segunda',
    1: 'Terça',
    2: 'Quarta',
    3: 'Quinta',
    4: 'Sexta',
    5: 'Sábado',
    6: 'Domingo'
}

lista_meses = [
    'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
    'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'
]
mapa_mes = {k: v for k,v in zip(range(1, 13), lista_meses)}

df['dia_da_semana'] = df.Data.dt.weekday.map(mapa_dia_semana)
df['Ano'] = df.Data.dt.year
df['Mês'] = df.Data.dt.month.map(mapa_mes)
df['semana'] = df.Data.dt.week

ordens = [mapa_dia_semana.values(), lista_meses, range(1,53)]

df['dia_da_semana'] = pd.Categorical(df['dia_da_semana'],
                                     categories=mapa_dia_semana.values(),
                                     ordered=True)

df['Mês'] = pd.Categorical(df['Mês'],
                           categories=lista_meses,
                           ordered=True)

df['final_de_semana'] = df.dia_da_semana.isin(['Sábado', 'Domingo'])

df['semana-dia_da_semana'] = df.semana.astype(str) + '-' + df.dia_da_semana.astype(str)
```
